```python
import pandas as pd #importing all the dependencies
```

```python
rawdata= pd.DataFrame({'Address':['45 road, kenith street, btm, bangalore, 50038', '45 road, street, hitech, hyd, 500038'],

'Date': [ '03-04-2022', '02-04-2022'],
'Name' : ['abc', 'dec'],
'Time': ['8:30:21', '5:30:21']})
```

```python
rawdata
```

```python
rawdata.head()
```

```python
rawdata.columns
```

```python
add_Split=rawdata.Address.str.split(',',expand=True)
```

```python
add_Split
```

```python
rawdata["RoadNo"] = add_Split[0]
rawdata["StreetName"] = add_Split[1]
rawdata["Area"] = add_Split[2]
rawdata["City"] = add_Split[3]
rawdata["Pincode"] = add_Split[4]
```

```python
rawdata
```

```python
date_Split=rawdata.Date.str.split('-', expand=True)
```

```python
date_Split
```

```python
rawdata["Day"]=date_Split[0]
rawdata["Month"]=date_Split[1]
rawdata["Year"]= date_Split[2]
rawdata
```

```python
time_Split=rawdata.Time.str.split(':',expand=True)
time_Split
```

```python
rawdata["Hour"] = time_Split[0]
rawdata["Minute"] = time_Split[1]
rawdata["Seconds"] = time_Split[2]
rawdata
```

```python
rawdata=rawdata.drop("Address", axis=1)
rawdata
```

```python
rawdata=rawdata.drop("Time",axis=1)
rawdata
```
